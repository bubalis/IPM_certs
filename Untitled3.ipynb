```python
def remove_leading_0(string):
    while string[0]=='0':
        string=string[1:]
    return string
```

```python
nums=[remove_leading_0(s).replace(',', '').strip() for s in pc_codes]
```

```python
def checker(pc_codes):
    pc_codes=str(pc_codes).split(', ')
    pc_codes=[remove_leading_0(s).replace(',', '').strip() for s in pc_codes]
    return any([pc_code in [str(id) for id in epa['Alternate ID'].unique()] for pc_code in pc_codes])
```
